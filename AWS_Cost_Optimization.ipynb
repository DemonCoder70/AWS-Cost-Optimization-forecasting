```python
#Read the Excel file into a DataFrame
df = pd.read_excel(filename)
print(df.head())
```

```python
id_vars = ['Accounts', 'Portfolio']  # keep these as identifiers
value_vars = [col for col in df.columns if col not in id_vars]  # all the date columns
```

```python
# Melt the wide table into long format
df_long = df.melt(
    id_vars=id_vars,
    value_vars=value_vars,
    var_name='Date',
    value_name='Cost'
)
```

```python
# Convert Date column to real datetime
df_long['Date'] = pd.to_datetime(df_long['Date'])
```

```python
# Sort for neatness
df_long = df_long.sort_values(['Date', 'Accounts']).reset_index(drop=True)
```

```python
print("Long-format sample:")
display(df_long.head(10))
```

```python
# 1. Remove invalid rows (like Total or missing Portfolio)
df_long_clean = df_long.dropna(subset=['Portfolio']).copy()
df_long_clean = df_long_clean[df_long_clean['Accounts'].str.lower() != 'total']
```

```python
# 2. Sort in a consistent way
df_long_clean = df_long_clean.sort_values(
    by=['Portfolio', 'Accounts', 'Date'],
    ascending=[True, True, True]
).reset_index(drop=True)
```

```python
print("Cleaned & sorted sample:")
display(df_long_clean.head(15))
```

```python
# Aggregate total spend per month
monthly_total = (
    df_long_clean
    .groupby('Date')['Cost']
    .sum()
    .sort_index()
)
```

```python
print("Monthly total spend:")
display(monthly_total)
```

```python
# Define training and testing windows
train_start = '2024-07-01'
train_end   = '2024-09-01'
test_start  = '2024-10-01'
test_end    = '2024-12-01'
```

```python
# Slice train & test series
train_series = monthly_total.loc[train_start:train_end]
test_series  = monthly_total.loc[test_start:test_end]
```

```python
print("TRAINING PERIOD:\n")
display(train_series)
```

```python
print("\nTEST PERIOD:\n")
display(test_series)
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import numpy as np
```

```python
# Fit Holt's Linear Trend model (no seasonality)
model = ExponentialSmoothing(
    train_series,
    trend='add',
    seasonal=None
)

fit = model.fit(optimized=True)
```

```python
# Forecast 3 months (matching your test period)
forecast = fit.forecast(3)
```

```python
print("FORECASTED VALUES:")
display(forecast)
```

```python
print("ACTUAL TEST VALUES:")
display(test_series)
```

```python
# Align forecast and actual values
results = pd.DataFrame({
    'Actual': test_series.values,
    'Forecast': forecast.values
}, index=test_series.index)
```

```python
# Calculate errors
results['Error'] = results['Actual'] - results['Forecast']
results['AbsError'] = results['Error'].abs()
results['PctError'] = results['AbsError'] / results['Actual'] * 100
```

```python
# Metrics
MAE = results['AbsError'].mean()
RMSE = np.sqrt((results['Error']**2).mean())
MAPE = results['PctError'].mean()
```

```python
print("RESULTS TABLE:")
display(results)
```

```python
print("\nACCURACY METRICS:")
print(f"MAE:  {MAE:,.2f}")
print(f"RMSE: {RMSE:,.2f}")
print(f"MAPE: {MAPE:.2f}%")
```

```python
# 1) Use the full monthly total series
full_series = monthly_total.copy()
```

```python
# Sanity check
print("Training from", full_series.index.min(), "to", full_series.index.max())
```

```python
# Forecast horizon: 5 months (Nov 2025  Mar 2026)
horizon = 5

last_date = full_series.index.max()
future_index = pd.date_range(
    start=last_date + pd.offsets.MonthBegin(1),
    periods=horizon,
    freq='MS'
)
```

```python
# -----------------------
# Model A: Holt Linear Trend (no seasonality)
# -----------------------
model_trend = ExponentialSmoothing(
    full_series,
    trend='add',
    seasonal=None
)
```

```python
fit_trend = model_trend.fit(optimized=True)
fc_trend = pd.Series(fit_trend.forecast(horizon), index=future_index)
```

```python
# -----------------------
# Model B: Holt-Winters (trend + yearly seasonality)
# We now have enough points (~18 months), so will try additive seasonality
# -----------------------
try:
    model_hw = ExponentialSmoothing(
        full_series,
        trend='add',
        seasonal='add',
        seasonal_periods=12
    )
    fit_hw = model_hw.fit(optimized=True)
    fc_hw = pd.Series(fit_hw.forecast(horizon), index=future_index)
except Exception as e:
    print("Seasonal model failed:", e)
    fc_hw = None
```

```python
# -----------------------
# Combine results in one table
# -----------------------
if fc_hw is not None:
    forecast_df = pd.DataFrame({
        'Forecast_Trend': fc_trend,
        'Forecast_HoltWinters': fc_hw
    })
else:
    forecast_df = pd.DataFrame({
        'Forecast_Trend': fc_trend
    })
```

```python
print("\nForecast for next 5 months (Nov 2025  Mar 2026):")
display(forecast_df)
```

```python
def classify_env(account):
    acc = account.lower()

    if "nonprod" in acc or "non-prod" in acc or "non_prod" in acc:
        return "nonprod"
    elif acc.endswith(".prod") or ".prod" in acc:
        return "prod"
    else:
        # No prod/nonprod tags; classify as other
        return "other"

df_long_clean['Env'] = df_long_clean['Accounts'].apply(classify_env)

df_long_clean.head(10)
```

```python
# Aggregate monthly spend by Portfolio and Environment
monthly_env_portfolio = (
    df_long_clean
    .groupby(['Date', 'Portfolio', 'Env'])['Cost']
    .sum()
    .reset_index()
)

print("Sample aggregated Portfolio  Env table:")
display(monthly_env_portfolio.head())
```

```python
monthly_env_pivot = monthly_env_portfolio.pivot_table(
    index='Date',
    columns=['Portfolio', 'Env'],
    values='Cost',
    aggfunc='sum'
).sort_index().fillna(0)

print("Monthly spend by Portfolio  Env (wide format):")
display(monthly_env_pivot.head(12))
```

```python
# Select trend window (highest-interest period)
trend_window = monthly_env_pivot.loc['2025-04-01':'2025-10-01']
print("Trend window preview:")
display(trend_window.head())
```

```python
slopes_env = {}

for (portfolio, env) in trend_window.columns:
    y = trend_window[(portfolio, env)].values

    # Skip empty or constant data
    if np.isnan(y).all() or np.allclose(y, 0):
        continue

    x = np.arange(len(y))
    slope, intercept = np.polyfit(x, y, 1)
    slopes_env[(portfolio, env)] = slope

trend_slopes_df = (
    pd.DataFrame.from_dict(slopes_env, orient='index', columns=['Monthly_Trend'])
    .sort_values('Monthly_Trend', ascending=False)
)

print("=== Portfolio  Env Trend Slopes (AprOct 2025) ===")
display(trend_slopes_df)
```

```python
# Filter for Vodapay only
vodapay_df = df_long_clean[df_long_clean["Portfolio"] == "Vodapay"].copy()

# Filter to AprilOct 2025 window
vodapay_df_window = vodapay_df[
    (vodapay_df["Date"] >= "2025-04-01") &
    (vodapay_df["Date"] <= "2025-10-01")
].copy()

display(vodapay_df_window.head())
```

```python
vodapay_pivot = vodapay_df_window.pivot_table(
    index="Date",
    columns=["Accounts", "Env"],
    values="Cost",
    aggfunc="sum"
).fillna(0)

display(vodapay_pivot)
```

```python
z_scores = (vodapay_pivot - vodapay_pivot.mean()) / vodapay_pivot.std()

anomalies_z = z_scores.abs() > 2.0   # using 2.0 threshold because short window

anomaly_points = vodapay_pivot[anomalies_z]

print("=== Z-Score Anomalies (AprOct 2025) ===")
display(anomaly_points.dropna(how="all"))
```

```python
mom_change = vodapay_pivot.diff()
mom_pct = vodapay_pivot.pct_change() * 100

# Flag: >25% growth OR >$1500 increase
mom_anomalies = (mom_pct > 25) | (mom_change > 1500)

print("=== MoM Anomalies (AprOct 2025) ===")
display(vodapay_pivot[mom_anomalies])
```

```python
def top_contributors(date):
    changes = mom_change.loc[date].sort_values(ascending=False)
    return changes[changes > 0].head(5)

print("=== Top Contributors for Anomaly Months ===")

for dt in mom_anomalies.index:
    if mom_anomalies.loc[dt].any():
        print(f"\n{dt.date()}  Top 5 Contributors:")
        display(top_contributors(dt))
```

```python
slopes = {}

for col in vodapay_pivot.columns:
    y = vodapay_pivot[col].values
    x = np.arange(len(y))

    if len(np.unique(y)) <= 1:
        slope = 0
    else:
        slope, _ = np.polyfit(x, y, 1)

    slopes[col] = slope

slopes_df = pd.DataFrame.from_dict(slopes, orient="index", columns=["MonthlyTrend"])
slopes_df = slopes_df.sort_values("MonthlyTrend", ascending=False)

print("=== Vodapay Accounts with Fastest Upward Trend (AprOct 2025) ===")
display(slopes_df.head(10))
```

```python
def analyze_portfolio_anomalies(portfolio_name,
                                start="2025-04-01",
                                end="2025-10-01",
                                z_thresh=2.0,
                                pct_thresh=25,
                                abs_thresh=1500):
    """
    Analyze per-account anomalies for one portfolio in a given date window.

    Returns:
      {
        'pivot': pivot_df,
        'z_anomalies': z_anomaly_df,
        'mom_anomalies': mom_anomaly_df,
        'top_contributors': top_contrib_dict,
        'slopes': slopes_df
      }
    """
    # 1) Filter portfolio + date window
    p_df = df_long_clean[df_long_clean["Portfolio"] == portfolio_name].copy()
    p_df = p_df[(p_df["Date"] >= start) & (p_df["Date"] <= end)]

    if p_df.empty:
        print(f"[{portfolio_name}] No data in range {start}  {end}.")
        return None

    # 2) Pivot: Date  (Account, Env)
    pivot = p_df.pivot_table(
        index="Date",
        columns=["Accounts", "Env"],
        values="Cost",
        aggfunc="sum"
    ).fillna(0)

    # 3) Z-score anomalies
    z_scores = (pivot - pivot.mean()) / pivot.std()
    z_anom_mask = z_scores.abs() > z_thresh
    z_anomalies = pivot.where(z_anom_mask).dropna(how="all")

    # 4) MoM anomalies: large growth
    mom_change = pivot.diff()
    mom_pct = pivot.pct_change() * 100

    mom_anom_mask = (mom_pct > pct_thresh) | (mom_change > abs_thresh)
    mom_anomalies = pivot.where(mom_anom_mask).dropna(how="all")

    # 5) Top contributors per anomaly month (based on MoM change)
    top_contributors = {}
    for dt in mom_anomalies.index:
        # MoM change at that date
        changes = mom_change.loc[dt].sort_values(ascending=False)
        top = changes[changes > 0].head(5)
        if not top.empty:
            top_contributors[dt] = top

    # 6) Trend slopes (MonthlyTrend) per account
    slopes = {}
    x = np.arange(len(pivot.index))
    for col in pivot.columns:
        y = pivot[col].values
        if len(np.unique(y)) <= 1:
            slope = 0.0
        else:
            slope, _ = np.polyfit(x, y, 1)
        slopes[col] = slope

    slopes_df = (
        pd.DataFrame.from_dict(slopes, orient="index", columns=["MonthlyTrend"])
        .sort_values("MonthlyTrend", ascending=False)
    )

    return {
        "pivot": pivot,
        "z_anomalies": z_anomalies,
        "mom_anomalies": mom_anomalies,
        "top_contributors": top_contributors,
        "slopes": slopes_df,
    }
```

```python
portfolios = sorted(df_long_clean["Portfolio"].unique())
all_results = {}
summary_rows = []

for p in portfolios:
    print(f"\n===== {p} =====")
    res = analyze_portfolio_anomalies(p)
    if res is None:
        continue
    all_results[p] = res

    # Top 3 upward-trending accounts per portfolio
    top3 = res["slopes"].head(3).copy()
    for (account, env), row in top3.iterrows():
        summary_rows.append({
            "Portfolio": p,
            "Account": account,
            "Env": env,
            "MonthlyTrend": row["MonthlyTrend"]
        })

summary_df = pd.DataFrame(summary_rows).sort_values(
    ["Portfolio", "MonthlyTrend"], ascending=[True, False]
)

print("\n=== Top Upward-Trending Accounts Per Portfolio (AprOct 2025) ===")
display(summary_df)
```
